# Plot a histogram to visualize the distribution 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant numeric columns
numeric_columns = [
  'Cost',
  'Age',
]

# Check for missing values in the selected numeric columns
missing_values = data[numeric_columns].isnull().sum()
if missing_values.sum() > 0:
    print("Warning: Missing values found in the selected numeric columns. Please handle missing values.")
else:
    # No missing values, proceed with the pair plot
    data_cleaned = data[numeric_columns]
    
    # Create the pair plot for numeric columns
    sns.pairplot(data_cleaned, height=2)
    plt.suptitle('Pairwise Relationships for Selected Numerical Columns')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Geography', palette='Set1')
plt.xlabel('Geography')
plt.title('Distribution of Data by Geography')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Age', palette='Set1')
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (assuming you've already loaded the dataset)
# Replace 'data' with the variable containing your dataset if it's different
data = pd.read_csv('CrimeSolutions.gov_Programs.csv')

# Create a new DataFrame for the counts of each combination of "Gender" and "Evidence Rating"
gender_evidence_counts = data.groupby(['Gender', 'Evidence Rating']).size().unstack(fill_value=0)

# Create a pie chart for each gender category
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen']

for i, gender in enumerate(gender_evidence_counts.index):
    ax = axes[i]
    ax.pie(gender_evidence_counts.loc[gender], labels=gender_evidence_counts.columns, colors=colors, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Distribution for {gender}')
    ax.axis('equal')

plt.suptitle('Distribution of Crime Solutions by Gender and Evidence Rating', fontsize=14)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (assuming you've already loaded the dataset)
# Replace 'data' with the variable containing your dataset if it's different
data = pd.read_csv('CrimeSolutions.gov_Programs.csv')

# Create a new DataFrame for the counts of each combination of "Gender" and "Evidence Rating"
gender_evidence_counts = data.groupby(['Gender', 'Evidence Rating']).size().unstack(fill_value=0)

# Create a pie chart for each gender category
fig, axes = plt.subplots(1, len(gender_evidence_counts), figsize=(12, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen']

for i, gender in enumerate(gender_evidence_counts.index):
    ax = axes[i]
    ax.pie(gender_evidence_counts.loc[gender], labels=gender_evidence_counts.columns, colors=colors, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Distribution for {gender}')
    ax.axis('equal')

plt.suptitle('Distribution of Crime Solutions by Gender and Evidence Rating', fontsize=14)
plt.show()
```
